````python
## Creating Linked List for assignment questions ##
class Node:
    def __init__(self,data):
        self.data =data
        self.next = None

# Creating a LinkedList with O(n) time complexity

def takeInput():
    inputList = [int(i) for i in input().split()]
    head = None
    for currData in inputList:
        if currData == -1:
            break
        newNode = Node(currData)
        if head == None:
            head = newNode
            tail = newNode
        else: 
            tail.next = newNode
            tail = newNode
            
    return head

# Print a LinkedList

def printLL(head):
    while head is not None:
        print(str(head.data)+'->',end='')
        head = head.next
    print(None)
    
    
head = takeInput()
printLL(head)
````

````python
def deleteNodeRec(start, pos) :
    if pos < 0:
        return start
    
    if start is None:
        return None
    
    if pos == 0:
        return start.next
    
    start.next = deleteNodeRec(start.next,pos-1)
    return start
````

````python
def length(head) :
    temp = head
    count=0
    while temp is not None:
        count+=1
        temp = temp.next
    return count
    
def appendLastNToFirst(head, N) :
    total_size = length(head)
    if total_size < N:
        return head
    curr = head
    c = 1
#     while count < total_size - k and curr is not None:
    while c < total_size - N :
        curr = curr.next
        c+=1
    NewNode = curr
    while curr.next is not None:
        curr = curr.next
    curr.next = head
    head = NewNode.next
    NewNode.next = None
    printLL(head)
````
