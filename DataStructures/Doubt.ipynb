```python
from sys import stdin

def isEmpty(data):
    return len(data) == 0

def top(data):
    # if isEmpty(data):
    #     return -1
    return data[len(data)-1]

def countBracketReversals(inputString) :
    s = []
    count=0
    if len(inputString)%2 != 0:
        return -1
    else:
        for i in range(len(inputString)):
            if inputString[i] == '{':
                s.append(inputString[i])
                
            elif inputString[i] == '}' and not isEmpty(s):
                if top(s) == "{":
                    s.pop()
                else:
                    s.append(inputString[i])
            # elif inputString[i] == '}' and isEmpty(s):
            else:
                s.append(inputString[i])
            # else:
            #     continue
    while not isEmpty(s):
        c1 = s.pop()
        c2 = s.pop()
        if c1 == c2:
            count+=1
        else:
            count+=2
    return count


#main
print(countBracketReversals(stdin.readline().strip()))
```

```python
from sys import stdin
import queue

class Stack :

    def __init__(self):
        self.q1 = queue.Queue()
        self.q2 = queue.Queue()

    def push(self,data):
        self.q1.put(data) 

    def pop(self):
        if self.isEmpty() :
            return -1
        
        while self.q1.qsize() > 1:
            self.q2.put(self.q1.get())
            
        x = self.q1.get()
        
        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp
        
        return x

    def top(self):
        if self.isEmpty():
            return -1
        
        while self.q1.qsize()>1:
            self.q2.put(self.q1.get())
            
        x=self.q1.get()
        self.q2.put(x)
        
        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp
            
        return x

    def getSize(self):
        return self.q1.qsize()

    def isEmpty(self):
        return self.getSize() == 0

#main
q = int(stdin.readline().strip())

stack = Stack()

while q > 0 :

	inputs = stdin.readline().strip().split(" ")
	choice = int(inputs[0])

	if choice == 1 :
		data = int(inputs[1])
		stack.push(data)

	elif choice == 2 :
		print(stack.pop())

	elif choice == 3 :
		print(stack.top())

	elif choice == 4 : 
		print(stack.getSize())

	else :
		if stack.isEmpty() :
			print("true")
		else :
			print("false")

	q -= 1
```
