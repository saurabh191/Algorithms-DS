### Queues using Array

````python
class queueArray:
    def __init__(self):
        self.__arr = []
        self.__count = 0
        self.__front = 0
    
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize() == 0
    
    def enque(self,item):
        self.__arr.append(item)
        self.__count+=1
        
    def dequeue(self):
        if self.isEmpty():
            print("Hey! Queue is already Empty!")
            return -1
        data = self.__arr[self.__front]
        self.__front +=1
        self.__count -=1
        return data
    
    def front(self):
        if self.isEmpty():
            print("Hey! Queue is already Empty!")
            return -1
        data = self.__arr[self.__front]
        return data
    

        
````

### Queues using LinkinedList

````python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class QueueLL:
    def __init__(self):
        self.__head = None
        self.__tail = None
        self.__count = 0
        
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize() == 0
    
    def enqueue(self,item):
        NewNode = Node(item)
        if self.__head is None:
            self.__head = NewNode
            self.__tail = NewNode
        else:
            self.__tail.next = NewNode
            self.__tail = NewNode
        self.__count +=1
        
    def dequeue(self):
        if self.isEmpty():
            print("Hey! Queue is already Empty!")
            return -1
        data = self.__head.data
        self.__head = self.__head.next
        self.__count -=1
        return data
    
    def front(self):
        if self.isEmpty():
            print("Hey! Queue is already Empty!")
            return -1
        data = self.__head.data
        return data   
    
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize() == 0
        
        
    
````
