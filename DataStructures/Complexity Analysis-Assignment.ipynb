- Code to find x to the power n

```python
def power_to(x,n):
    if n==0:
        return 1
    small_ouput = power_to(x,n//2)
    if n%2==0:
        return small_ouput * small_ouput
    else:
        return x * small_ouput * small_ouput
```

```python
from sys import stdin

# Partition function for array sorting
def Partition(arr,start,end):
    pivot = arr[start]
    count = 0
    for i in range(start,end+1):
        if arr[i] < pivot:
            count+=1
    PI = start+count
    arr[start],arr[count+start] = arr[count+start],arr[start]
    i=start
    j=end
    while i<j:
        if arr[i]<pivot:
            i+=1
        elif arr[j]>pivot:
            j-=1
        else:
            arr[i],arr[j]=arr[j],arr[i]
            i+=1
            j-=1
    return PI

# Sorting the arrays
def sortArray(arr,start,end):
    if start >= end:
        return
    pivot_index = Partition(arr,start,end)
    sortArray(arr,start,pivot_index-1)
    sortArray(arr,pivot_index+1,end)
    return arr
    
    
def intersection(arr1, arr2, n, m):
    sortArray(arr1,0,len(arr1)-1)
    sortArray(arr2,0,len(arr2)-1)
    i=0
    j=0
    k=0
    # arr=[]
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            i+=1
        elif arr1[i] > arr2[j]:
            j+=1
        else:
            print(arr1[i],end=" ")
            i+=1
            j+=1
            k+=1
    # print(*arr)

# Taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    
    if n == 0 :
    	return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


#main
t = int(float(stdin.readline().strip()))

while t > 0 :

    arr1, n = takeInput()
    arr2, m = takeInput()
    intersection(arr1, arr2, n, m)
    print()
    t -= 1
```

- Array Equilibrium Index

```python
#Approach-1
#
#Time complexity will be O(n^2)
#

from sys import stdin

def arrayEquilibriumIndex(arr, n) :
    if len(arr) == 1:
        return 0
    if len(arr) == 0:
        return -1
    # sum_left = 0
    # sum_right = 0
    for i in range(1,len(arr)):
        sum_left = sum(arr[0:i])
        sum_right = sum(arr[i+1:len(arr)])
        if sum_left == sum_right:
            return i
    return -1
 
#Taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :
    
    arr, n = takeInput()
    print(arrayEquilibriumIndex(arr, n))

    t-= 1
```

- Find pair sum in a array

```python
from sys import stdin

def pairSum(arr, n, num) :
    value=0
    count=0
    mydict = dict()
    for i in range(n):
        temp = num - arr[i]
        if temp in mydict:
            value = mydict[temp]
            for j in range(value):
                count+=1
        if arr[i] in mydict:
            mydict[arr[i]] +=1
        else:
            mydict[arr[i]] = 1
    return count

#taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :
    
    arr, n = takeInput()
    num = int(stdin.readline().strip())
    print(pairSum(arr, n, num))

    t -= 1
```

- Rotate array by 'D' elements

```python
#Time complexity = O(n^2)

def rotate(arr, n, d):
    for i in range(d):
        temp = arr[0]
        for j in range(len(arr)-1):
            arr[j] = arr[j+1]
        arr[-1] = temp
    return arr
```

```python
# Time complexity = O(n)
def rotateBetter(arr, n, d):
    temp = arr[0:d]
    for j in range(len(arr[d:n])):
        arr[j] = arr[j+d]
    arr[-d:]=temp
    return arr
```

```python
from sys import stdin


def tripletSum(arr, n, num) :
    arr.sort()
    count = 0
    for i in range(n):
        target = num-arr[i]
        count+=pairSum(arr,i+1,n-1,target)
    return count

def pairSum(arr,s,n,num):
    startIndex = s
    endIndex = n
    numPair = 0
    while startIndex < endIndex:
        if arr[startIndex] + arr[endIndex] < num:
            startIndex+=1
        elif arr[startIndex] + arr[endIndex] > num:
            endIndex-=1
        else:
            elementAtStart = arr[startIndex]
            elementAtEnd = arr[endIndex]
            
            if elementAtStart == elementAtEnd:
                totalElementsFromStartToEnd = (endIndex - startIndex) +1
                numPair += ((totalElementsFromStartToEnd)*(totalElementsFromStartToEnd -1)//2)
                return numPair
            
            tempStartIndex = startIndex+1
            tempEndIndex = endIndex-1
            while (tempStartIndex <= tempEndIndex) and (arr[tempStartIndex] == elementAtStart):
                tempStartIndex+=1
            while (tempEndIndex >= tempStartIndex) and (arr[tempEndIndex] == elementAtEnd):
                tempEndIndex-=1
            totalElementsFromStart = (tempStartIndex - startIndex)
            totalElementsFromEnd = (endIndex - tempEndIndex)
            numPair += (totalElementsFromStart * totalElementsFromEnd)
            startIndex = tempStartIndex
            endIndex = tempEndIndex
    return numPair

#taking input using fast I/O method
def takeInput() :
    n = int(stdin.readline().strip())
    if n == 0 :
        return list(), 0

    arr = list(map(int, stdin.readline().strip().split(" ")))
    return arr, n


def printList(arr, n) : 
    for i in range(n) :
        print(arr[i], end = " ")
    print()


#main
t = int(stdin.readline().strip())

while t > 0 :
    
    arr, n = takeInput()
    num = int(stdin.readline().strip())
    print(tripletSum(arr, n, num))

    t -= 1
```
